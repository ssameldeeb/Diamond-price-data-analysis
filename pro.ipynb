```python
import pandas as pd 
import datetime as dt
import time
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data = pd.read_csv("DiamondsPrices.csv")
```

```python
print(data.shape)
print(data.dtypes)
print(data.isna().sum())
print(data.columns.values)
print(data.head())
```

```python
print(data["color"].value_counts())
```

```python
print(data[data['color'].str.contains('G')].head())
print(data[data['color'].isin(['G'])].head())
```

```python
for x in data.columns.values:
    if data[x].dtype == "object":
        print(x.upper())
        print(dt.datetime.now().strftime("%H:%M:%S"))
        print(data[x].value_counts())
        print(data[x].describe())
        print("_"*100)
        
        plt.figure(figsize=(10,5))
        sns.countplot(data[x])
        plt.tight_layout()
        plt.savefig("save/"+x+".png")
        plt.show()
        
    else:
        print(x.upper())
        print(dt.datetime.now().strftime("%H:%M:%S"))
        print(data[x].value_counts())
        print(data[x].describe())
        print("_"*100)
        
        plt.figure(figsize=(10,5))
        sns.distplot(data[x])
        plt.tight_layout()
        plt.savefig("save/"+x+".png")
        plt.show()
        
        
```
